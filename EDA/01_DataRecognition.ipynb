**Correlation Matrix**
---

---

```python
plt.figure(figsize=(18,12))
ax = plt.axes()
matrix = np.triu(df.corr())
sns.heatmap(df.corr(), linewidths=.5, fmt= '.1f',ax=ax, mask=matrix,cmap="YlGnBu")
plt.plot()
```

---
**Box Plots** 
---

---

```python
fig, axes = plt.subplots(10, 3, figsize=(20, 60))
axes = axes.reshape(30,)

for i,column in enumerate(df.columns[:-1]):
    axes[i].set_yscale('symlog')
    sns.boxplot(ax = axes[i],x="Class", y=column,data=df, palette="Set3")
```
